```python
## Data Visualization
### Sales Trend Over Time
```

```python
import matplotlib.pyplot as plt

df['Order Date'] = pd.to_datetime(df['Order Date'])

sales_trend = df.groupby('Order Date')['Sales'].sum()

sales_trend.plot()
plt.title("Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```

### Top 10 Products

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

top_products.plot(kind='bar')
plt.title("Top 10 Products by Sales")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()
```

### Sales by Region

```python
region_sales = df.groupby('Region')['Sales'].sum()

region_sales.plot(kind='bar')
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```

### Category Analysis

```python
category_sales = df.groupby('Category')['Sales'].sum()

category_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title("Sales by Category")
plt.show()
```

## Key Insights

## Key Insights
- Sales fluctuate over time with noticeable peaks and drops.
- A small number of products contribute to a large portion of total revenue.
- The West region generates the highest sales among all regions.
- The Technology category contributes the most to overall revenue.
